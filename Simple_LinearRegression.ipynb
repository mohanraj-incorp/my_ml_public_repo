```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, r2_score, mean_squared_error
from sklearn.model_selection import train_test_split
import statsmodels
import statsmodels.api as sm
# supress warning
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
adv_df = pd.read_csv(r"datasets\LinReg_Advertising.csv")
```

```python
adv_df=adv_df.drop('Unnamed: 0', axis=1)
```

```python
adv_df.head(5)
```

```python
adv_df.info()
```

```python
sns.pairplot(data=adv_df,x_vars=['TV','Radio','Newspaper'], y_vars='Sales',size=4, aspect=1, kind='scatter')
plt.show()
```

```python
sns.heatmap(adv_df.corr(), annot=True, cmap='YlGnBu' )
```

```python
x = adv_df[['TV','Radio']]
y=adv_df['Sales']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, train_size=0.7, random_state=100)
```

```python
LR_model =LinearRegression()
LR_model.fit(x_train,y_train)
LR_model
```

```python
y_train_pred = LR_model.predict(x_train)

y_test_pred = LR_model.predict(x_test)

print(f"Training data \n Mean squared error: {mean_squared_error(y_train, y_train_pred):.2f}")

# The coefficient of determination (R-squared)
print(f"Coefficient of determination (R-squared): {r2_score(y_train, y_train_pred):.2f}")

print(f"Test data \n Mean squared error: {mean_squared_error(y_test, y_test_pred):.2f}")

# The coefficient of determination (R-squared)
print(f"Coefficient of determination (R-squared): {r2_score(y_test, y_test_pred):.2f}")
```

```python
x = adv_df['TV']
y=adv_df['Sales']

x_train, x_test, y_train, y_test = train_test_split(x,y, train_size=0.7, random_state=100)

x_train_reshape=x_train.values.reshape(-1,1)
x_test_reshape = x_test.values.reshape(-1,1)
```

```python
LR_model =LinearRegression()
LR_model.fit(x_train_reshape,y_train)
LR_model
```

```python
y_train_pred = LR_model.predict(x_train_reshape)

y_test_pred = LR_model.predict(x_test_reshape)

print(f"Training data \n Mean squared error: {mean_squared_error(y_train, y_train_pred):.2f}")

# The coefficient of determination (R-squared)
print(f"Coefficient of determination (R-squared): {r2_score(y_train, y_train_pred):.2f}")

print(f"Test data \n Mean squared error: {mean_squared_error(y_test, y_test_pred):.2f}")

# The coefficient of determination (R-squared)
print(f"Coefficient of determination (R-squared): {r2_score(y_test, y_test_pred):.2f}")
```
